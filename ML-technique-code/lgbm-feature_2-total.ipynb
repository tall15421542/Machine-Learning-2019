```python
import lightgbm as lgb
import numpy as np
```

```python
feature = 2
```

```python
train_X = np.load('../X_train.npz')
x_train = train_X['arr_0']

test_X = np.load('../X_test.npz')
x_pred = test_X['arr_0']


train_Y = np.load('../Y_train.npz')
y_train = (train_Y['arr_0'])

pen_rate = y_train[:,feature]

feature_size = 500

rf_index = []
with open("rf_index_2.csv") as f:
    for line in f:
        rf_index.append(int(line.strip()))
rf_index = np.array(rf_index)

x_train = x_train[:,rf_index[:feature_size]]


x_pred = x_pred[:,rf_index[:feature_size]]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_train, pen_rate, test_size=0.05, random_state=42)
```

```python
lgb_train = lgb.Dataset(x_train,pen_rate)
#lgb_eval = lgb.Dataset(x_test,y_test,reference=lgb_train)
```

```python
param = {
        'num_leaves': 500,
        'max_bin': 255,
        'min_data_in_leaf': 1,
        'learning_rate': 0.1,
        'bagging_fraction': 1,
        'bagging_freq': 2,
        'feature_fraction': 1,
        #'min_gain_to_split': 0.65,
        'max_depth': 15,
        'save_binary': True,
        'seed': 1337,
        'feature_fraction_seed': 1337,
        'bagging_seed': 1337,
        'data_random_seed': 1337,
        'objective': 'regression_l1',
        'boosting_type': 'gbdt',
        #'lambda_l1':5,
        #'boosting_type': 'dart',

        'verbose': 0,
        'metric': {'l1'},
        'device': 'gpu',
        'gpu_platform_id': 0,
        'gpu_device_id': 0
    }
```

```python
gbm = lgb.train(param,lgb_train,num_boost_round=2500,valid_sets=[lgb_train])#,callbacks=[lgb.reset_parameter(learning_rate=[0.1]*1000+[0.02]*1000)])#,early_stopping_rounds=10)#,callbacks=[lgb.reset_parameter(learning_rate=[0.02]*5000+[0.002]*5000)])
```

```python
y1_pred = gbm.predict(x_pred,num_iteration=gbm.best_iteration)
```

```python
for i in range(len(y1_pred)):
    if y1_pred[i]<0.5:
        y1_pred[i] = 0.5
    if y1_pred[i]>1.0:
        y1_pred[i]= 1.0
        
```

```python
y_feature = 'y'+str(feature)+'.csv'
y_file = open(y_feature,"w")
for i in range(len(y1_pred)):
    y_file.write(str(y1_pred[i]))
    if i != len(y1_pred)-1:
        y_file.write("\n")
y_file.close()
```
